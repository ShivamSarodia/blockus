```python
data = []

with open("/Users/shivamsarodia/Downloads/detail.txt") as f:
    lines = f.readlines()
    for line in lines:
        if "Precinct" in line:
            all_headers = line.split("  ")
            headers = [x.strip() for x in all_headers if x.strip()]
            for i in range(2, 7):
                headers[i] = "D - " + headers[i]
            for i in range(7, 12):
                headers[i] = "R - " + headers[i]
        else:
            all_data = line.split("  ")
            this_data = [int(x.strip()) for x in all_data if x.strip()]
            
            assert len(this_data) == len(headers)

            data.append({headers[i]: this_data[i] for i in range(len(headers))})
```

```python
fraction_dem = []

for d in data:
    if d["Total"] < 100:
        continue

    early_vote_d = d["D - Early Voting"] + d["D - Vote by Mail"]
    early_vote_r = d["R - Early Voting"] + d["R - Vote by Mail"]

    fraction_dem.append(early_vote_d / (early_vote_r + early_vote_d))
```

```python
import matplotlib.pyplot as plt

plt.hist(fraction_dem, bins=10, range=(0, 1), edgecolor='black')
plt.title('Histogram of Democratic Vote Fractions')
plt.xlabel('Fraction of Democratic Votes')
plt.ylabel('Number of Precincts')
plt.show()
```
